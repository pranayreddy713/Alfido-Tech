Import the **libraries** and **packages**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

Using **sep="::"** we seperate the data into different columns

```python
df_movie=pd.read_csv('/content/movies.dat',sep="::",engine='python')
df_movie.dropna(inplace=True)
df_movie.head()
```

```python
column_names = ['MovieIDs', 'MovieName', 'Genre']
df_movie.columns = column_names
```

Downloading ratings dataset

```python
df_ratings=pd.read_csv("/content/ratings.dat", sep="::", engine="python")
df_ratings.dropna(inplace=True)
df_ratings.head(10)
```

There is no Column names so we should provide it

```python
column_names = ['UserId', 'MovieID', 'Ratings','Unknown']
df_ratings.columns = column_names
```

Loading the User Data given by alfido

```python
df_users=pd.read_csv("/content/users.dat", sep="::", engine="python")
df_users.dropna(inplace=True)
df_users.head(10)
```

Again there is no mention of column names so we should provide it

```python
column_names = ['UserID', 'Gender', 'Age','Occupation','Zipcode']
df_users.columns = column_names
```

Label encoding is used to convert categorical data into a format that **machine** **learning** **algorithms** can work

```python
labelencoder = LabelEncoder()

df_users['Gender']= labelencoder.fit_transform(df_users['Gender'])

df_users.head()
```

Concatenating the datasets for combining the 3 datasets of movies,users and rating

```python
df_data=pd.concat([df_movie,df_ratings,df_users],axis=1)
df_data.head(10)
```

To find out how many rows and columns are in out concatinated datset

```python
df_data.shape
```

Removing the unwanted columns

```python
df2=df_data.drop(["Occupation","Zipcode","Unknown"],axis=1)
df2.head()
```

To check the missing values in the dataframe2(df2)

```python
df2.isna().sum()
```

Handling the missing values

```python
df_final=df2.dropna()
```
